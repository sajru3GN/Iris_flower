```python
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 4)
x_train
```

```python
from sklearn.svm import SVC
svn = SVC()
svn.fit(x_train,y_train)
```

```python
predictions = svn.predict(x_test)
accuracy_score(y_test,predictions)
```
